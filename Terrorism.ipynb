##### 2.3.1 Using multiple API's to clean State, Latitude and Longitude

##### This part will be long since there are many nulls and the API's am using is a free version so it has a daliy limit of requests

```python
import ssl
from geopy.geocoders import Nominatim

# Set SSL certificate verification to false
ssl._create_default_https_context = ssl._create_unverified_context

# Create a geocoder object using GeoNames
geolocator = Nominatim(user_agent='my_app')

# Filter rows with null values in 'State' column
null_rows = df1[df1['State'].isnull()]

# Iterate over the null rows
for index, row in null_rows.iterrows():
    try:
        # Reverse geocode the latitude and longitude values to get the location information
        location = geolocator.reverse(
            (row['Latitude'], row['Longitude']),
            exactly_one=True,
            language='en'
        )

        # If location is found, extract the state value
        if location:
            state = location.raw.get('address', {}).get('state')

            # If state is None, try extracting other location details
            if state is None:
                city = location.raw.get('address', {}).get('city')
                town = location.raw.get('address', {}).get('town')

                # Prioritize the available location details
                state = city or town 

            # Assign the state value to the DataFrame
            df1.at[index, 'State'] = state

            # Print the filled value
            print(f"Filled null value in 'State' column at index {index}. State: {state}")
        else:
            print(f"No location found for latitude: {row['Latitude']}, longitude: {row['Longitude']}")

    except Exception as e:
        print(f"An error occurred for index {index}: {str(e)}")
```

```python
import ssl
from geopy.geocoders import Nominatim

# Set SSL certificate verification to false
ssl._create_default_https_context = ssl._create_unverified_context

# Create a geocoder object using GeoNames
geolocator = Nominatim(user_agent='my_app')

# Filter rows with null values in 'Latitude' or 'Longitude' columns
null_rows = df1[df1['Latitude'].isnull() | df1['Longitude'].isnull()]

# Iterate over the null rows
for index, row in null_rows.iterrows():
    try:
        # Get the state value
        state = row['State']

        # Skip if the state value is also null
        if pd.isnull(state):
            continue

        # Geocode the state to get the latitude and longitude values
        location = geolocator.geocode(state, exactly_one=True, language='en')

        # If location is found, extract the latitude and longitude
        if location:
            latitude = location.latitude
            longitude = location.longitude

            # Assign the latitude and longitude values to the DataFrame
            df1.at[index, 'Latitude'] = latitude
            df1.at[index, 'Longitude'] = longitude

            # Print the filled values
            print(f"Filled null values in 'Latitude' and 'Longitude' columns at index {index}. Latitude: {latitude}, Longitude: {longitude}")
        else:
            print(f"No location found for state: {state}")

    except Exception as e:
        print(f"An error occurred for index {index}: {str(e)}")
```
